## Trend Forecasting with a Linear Model

```python
def window_dataset(series, window_size, batch_size=32,
                   shuffle_buffer=1000):
    dataset = tf.data.Dataset.from_tensor_slices(series)
    dataset = dataset.window(window_size + 1, shift=1, drop_remainder=True)
    dataset = dataset.flat_map(lambda window: window.batch(window_size + 1))
    dataset = dataset.shuffle(shuffle_buffer)
    dataset = dataset.map(lambda window: (window[:-1], window[-1]))
    dataset = dataset.batch(batch_size).prefetch(1)
    return dataset
```

```python
def model_forecast(model, series, window_size):
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size))
    ds = ds.batch(32).prefetch(1)
    forecast = model.predict(ds)
    return forecast
```

### Preprocessing With 1D-Convolutional Layers

```python
def seq2seq_window_dataset(series, window_size, batch_size=32,
                           shuffle_buffer=1000):
    series = tf.expand_dims(series, axis=-1)
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    ds = ds.shuffle(shuffle_buffer)
    ds = ds.map(lambda w: (w[:-1], w[1:]))
    return ds.batch(batch_size).prefetch(1)
```

```python
# Step 1: Convert 'creation_date' to datetime
data['creation_date'] = pd.to_datetime(data['creation_date'])

# Step 2: Filter the dataset for selected tags (python, machine-learning, deep-learning)
# To achieve this, we need to check if these tags are present in the 'tags' column which contains tags in a list format as strings

# The tags column needs to be converted from string representation of lists to actual lists
import ast  # Importing ast to safely evaluate string literals as python expressions

data['tags'] = data['tags'].apply(ast.literal_eval)

# Define the tags of interest
tags_of_interest = ['python', 'machine-learning', 'deep-learning']

# Filter the data to include only rows where the tags column contains any of the tags of interest
filtered_data = data[data['tags'].apply(lambda tags: any(tag in tags_of_interest for tag in tags))]

# Step 3: Create a time series of counts per day for each tag
# Initialize a dataframe to store the counts
time_series_data = pd.DataFrame()

for tag in tags_of_interest:
    # Filter data for each tag
    tag_data = filtered_data[filtered_data['tags'].apply(lambda tags: tag in tags)]
    # Group by date and count the number of posts
    daily_counts = tag_data.groupby(tag_data['creation_date'].dt.date).size().reset_index(name=f'count_{tag}')
    # Merge the counts into the time_series_data dataframe
    if time_series_data.empty:
        time_series_data = daily_counts
    else:
        time_series_data = pd.merge(time_series_data, daily_counts, how='outer', on='creation_date')

# Fill missing values with 0 (days with no posts)
time_series_data = time_series_data.fillna(0)

# Show a summary of the created time series data
time_series_data.head(), time_series_data.info()
```
